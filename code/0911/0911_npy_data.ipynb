```python
import numpy as np
import pandas as pd
import re

samsung = pd.read_csv('samsung0910.csv')
samsung = samsung[::-1]
samsung.reset_index(drop=True, inplace=True)
samsung
```

```python
samsung.isnull().sum()
```

```python
samsung['시가'] = samsung['시가'].map(lambda x: int(re.sub(r',', '', x)))
samsung['고가'] = samsung['고가'].map(lambda x: int(re.sub(r',', '', x)))
samsung['저가'] = samsung['저가'].map(lambda x: int(re.sub(r',', '', x)))
samsung['종가'] = samsung['종가'].map(lambda x: int(re.sub(r',', '', x)))
samsung['거래량'] = samsung['거래량'].map(lambda x: int(re.sub(r',', '', x)))
samsung
```

```python
x = samsung[['시가','고가','저가','종가','거래량']].values
y = samsung['종가'].values
#y = samsung['종가'].shift(20).values
```

```python
np.save('x_data.npy', arr=x)
np.save('y_data.npy', arr=y)
```
